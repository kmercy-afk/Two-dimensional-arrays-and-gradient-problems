```python
import numpy as np

# Define x and y arrays
x = np.arange(-50, 50.1, 0.1)  # x values from -50 to 50 with 0.1 intervals
y = (1/2) * x + 1             # y values based on the equation y = 1/2 * x + 1

# Compute differences in y (dy) and x (dx)
dy = y[1:] - y[:-1]  # Differences in y
dx = x[1:] - x[:-1]  # Differences in x (this will be constant as x is uniformly spaced)

# Compute gradient (dy/dx)
gradient = dy / dx

# Display results
print("Gradients:", gradient)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create x ndarray with values from -50 to 50 with 0.1 intervals
x = np.arange(-50, 50.1, 0.1)

# Calculate y for each x using the equation y = (1/2)x + 1
y = (1/2) * x + 1

# Calculate the difference in y and x between adjacent elements
dy = np.diff(y)  # Change in y
dx = np.diff(x)  # Change in x

# Compute the gradient as dy/dx
gradient = dy / dx  # Element-wise division

# Midpoints for x values (for gradient plot)
x_midpoints = x[:-1] + dx / 2

# Plot the linear function
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)  # First plot
plt.plot(x, y, label="y = (1/2)x + 1", color="blue")
plt.title("Linear Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()

# Plot the gradient
plt.subplot(1, 2, 2)  # Second plot
plt.plot(x_midpoints, gradient, label="Gradient (dy/dx)", color="red")
plt.title("Gradient of Linear Function")
plt.xlabel("x")
plt.ylabel("Gradient")
plt.grid()
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_gradients_and_plot(equation, x_range, x_interval, title):
    """
    Calculates the y values and gradients for a given equation, and plots the results.
    
    Args:
        equation (callable): A function defining y = f(x).
        x_range (tuple): Range of x values as (start, stop).
        x_interval (float): Interval between consecutive x values.
        title (str): Title for the graphs.
        
    Returns:
        tuple: (ndarray of x values, ndarray of y values, ndarray of gradients)
    """
    # Step 1: Create x ndarray
    x = np.arange(x_range[0], x_range[1] + x_interval, x_interval)
    
    # Step 2: Calculate y ndarray
    y = equation(x)
    
    # Step 3: Calculate gradients
    dy = np.diff(y)  # Change in y
    dx = np.diff(x)  # Change in x (all will be the same, x_interval)
    gradient = dy / dx  # Element-wise division
    
    # Midpoints of x for gradient plotting
    x_midpoints = x[:-1] + dx / 2
    
    # Plotting
    plt.figure(figsize=(12, 5))
    
    # Plot the equation
    plt.subplot(1, 2, 1)
    plt.plot(x, y, label=title, color="blue")
    plt.title(f"Graph of {title}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid()
    plt.legend()
    
    # Plot the gradient
    plt.subplot(1, 2, 2)
    plt.plot(x_midpoints, gradient, label=f"Gradient of {title}", color="red")
    plt.title(f"Gradient of {title}")
    plt.xlabel("x")
    plt.ylabel("Gradient")
    plt.grid()
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    return x, y, gradient

# Define the equations
def equation1(x):
    return x**2

def equation2(x):
    return 2 * x**2 + 2 * x

def equation3(x):
    return np.sin(np.sqrt(x))

# Calculate gradients and plot for each equation
x1, y1, gradient1 = calculate_gradients_and_plot(equation1, (-50, 50), 0.1, "y = x^2")
x2, y2, gradient2 = calculate_gradients_and_plot(equation2, (-50, 50), 0.1, "y = 2x^2 + 2x")
x3, y3, gradient3 = calculate_gradients_and_plot(equation3, (0, 50), 0.1, "y = sin(sqrt(x))")

# Output gradients as ndarrays
print("Gradients for y = x^2:", gradient1)
print("Gradients for y = 2x^2 + 2x:", gradient2)
print("Gradients for y = sin(sqrt(x)):", gradient3)
```

```python
import numpy as np

def find_minimum_and_gradients(equation, x_range, x_interval):
    """
    Finds the minimum value of y, its corresponding x, and the gradients before and after the minimum.

    Args:
        equation (callable): A function defining y = f(x).
        x_range (tuple): Range of x values as (start, stop).
        x_interval (float): Interval between consecutive x values.

    Returns:
        dict: A dictionary containing the minimum y, corresponding x, and gradients.
    """
    # Create x ndarray
    x = np.arange(x_range[0], x_range[1] + x_interval, x_interval)
    
    # Calculate y ndarray
    y = equation(x)
    
    # Find minimum y and its index
    y_min = y.min()
    index_min = y.argmin()
    x_min = x[index_min]
    
    # Compute gradients (dy/dx)
    dy = np.diff(y)
    dx = np.diff(x)
    gradient = dy / dx  # Element-wise division
    
    # Find gradients before and after the minimum
    gradient_before = gradient[index_min - 1] if index_min > 0 else None
    gradient_after = gradient[index_min] if index_min < len(gradient) else None
    
    # Return results
    return {
        "x_min": x_min,
        "y_min": y_min,
        "gradient_before": gradient_before,
        "gradient_after": gradient_after,
    }

# Define the equation from problem 5
def equation3(x):
    return np.sin(np.sqrt(x))

# Find the minimum value, corresponding x, and gradients for the given equation
results = find_minimum_and_gradients(equation3, (0, 50), 0.1)

# Print results
print("Minimum y:", results["y_min"])
print("Corresponding x:", results["x_min"])
print("Gradient before minimum:", results["gradient_before"])
print("Gradient after minimum:", results["gradient_after"])
```
